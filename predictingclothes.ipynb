```python
from __future__ import absolute_import, division, print_function, unicode_literals


import tensorflow as tf
```

```python
mnist = tf.keras.datasets.fashion_mnist

(x_train, y_train), (x_test, y_test) = mnist.load_data()
x_train, x_test = x_train / 255.0, x_test / 255.0
```

```python
models = tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape = (28,28)),
    tf.keras.layers.Dense(128, activation = "relu"),
    tf.keras.layers.Dense(10, activation = "softmax")
])
```

```python
models.compile(optimizer = 'adam', loss = "sparse_categorical_crossentropy", metrics = ["accuracy"])
```

```python
models.fit(x_train, y_train, epochs=5)
```

```python
prediction = models.predict(x_test)
```

```python
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
```

```python
lo,acc = models.evaluate(x_test,y_test)
print("TEST ACCURACY = ",acc)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
for i in range(10):
    print("Predicted : ",class_names[np.argmax(prediction[i])])
    print("Actual : ",class_names[y_test[i]])
    imgplot = plt.imshow(x_test[i])
    plt.show()
```
